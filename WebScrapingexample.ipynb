```
#Target Url
trend_page = 'https://finance.yahoo.com/trending-tickers'
```

Import Libraries


```
import pandas as pd
```

PANDA dataframes

```
tables = pd.read_html(trend_page)
len(tables)
```

```
trends = tables[0]
```

```
#get current date
from datetime import date
to_day=date.today()
print(to_day)
```

```
trends.insert(9,'DateCol',to_day,True)
```

```

trends.head(100)
```
